```python
import nltk
nltk.download('stopwords')
from pyresparser import ResumeParser
import os
from docx import Document
```

```python
##file format should be in .txt , .docx or .pdf only
filed=input()
```

```python
try:
    doc = Document()
    with open(filed, 'r') as file:
        doc.add_paragraph(file.read())
    doc.save("text.docx")
    data = ResumeParser('text.docx').get_extracted_data()
    print(data['skills'])
except:
    data = ResumeParser(filed).get_extracted_data()
    print(data['skills'])
```

```python
data
```

```python
import pandas as pd
```

```python
data1=pd.DataFrame.from_dict(data, orient='index')
data1
```

```python
#dict_df = pd.DataFrame({ key:pd.Series(value) for key, value in data.items() })
#dict_df
my_data = dict([ (k,pd.Series(v)) for k,v in data.items() ])
my_data
```

```python
import numpy as np
def f(x):
  values = x[~x.isnull()].values
  values = np.resize(values, len(x))
  return values
```

```python
newdf = pd.DataFrame(my_data)
newdf = newdf.apply(f, axis=0)
newdf
```

```python
from pathlib import Path  
filepath = Path('SyedAliShan.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
newdf.to_csv(filepath)
```
